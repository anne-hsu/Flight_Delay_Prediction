# 1) Abstract

Flight departure delays are costly, directly costing the U.S. airline industry approximately \$10 billion annually. These inefficiencies have indirect impacts (e.g. time loss, productivity losses across all sectors) that ripple throughout and cost the U.S. economy an estimated \$30-35 billion per year. Therefore, there is great incentive to addressing this problem. Because a missed delay (passenger stranded, crew duty-time pressure, downstream cascade) is materially more expensive than an extra advisory on a flight that ends up on time, we evaluate primarily with **F_β at β = 2**, which weights recall **twice** as heavily as precision. 

We present May(De)lay, a research project delving into the problem of delayed departure flights. We framed this problem as a binary classification task, focusing on predicting if a flight will be delayed by 15+ minutes one hour before departure time. The outcome variable is DEP_DEL15, where 1 indicates a 15+ minute departure delay and 0 indicates on-time or a shorter delay. Being able to identify and understand if and why flights are being delayed is important both for current operations and any further work building upon this project. For our model development, we leveraged two main datasets: (1) the Bureau of Transportation Statistics (BTS) flight records dataset and (2) the NOAA's Local Climatological Data (LCD) weather observations dataset (These will be referred to as flight and weather data from here on out). 

The scenario in which we perform binary classification is simple. For every flight, we want to predict if it'll be delayed by 15+ minutes one hour before departure time. We begin by rounding the departure time down to the nearest half hour (e.g. 10:24am --> 10am, meaning our prediction will be made at 9am). From there, we transform the most recent 3 hours of available weather data (6am - 9am) into a high signal vector, containing features we have determined to be important. This is the manner in which a time series dataset was constructed using the flight and weather datasets. The final dataset upon which models were trained spans five years (2015–2019) and consists of 31.7 million flight records

Among the classical models, GBT achieved the strongest validation performance (F2 = 0.523, AUROC = 0.695 on the 5-year data, at threshold 0.20), with all models achieving approximately 85–91% recall at the tuned decision threshold. The Phase 3 feature engineering pipeline improved AUROC by 1–2 percentage points over Phase 2 baselines across all model types. The MLP neural network, trained with a bespoke StandardScaler preprocessing pipeline and complemented by a scikit-learn ablation study examining Adam optimization, L2 regularization, and feature family contributions, represents our response to the client's requirement for a deep learning approach. Scale-up experiments from one year to five years of training data yielded modest but consistent improvements, confirming that additional historical context provides incremental predictive value.

## 2a) Weather Data: NOAA's Local Climatological Data (LCD)

### Data Wrangling

The original full weather dataset spanned 2015-2021. During 2020-2021, the world suffered from the COVID-19 pandemic. In regards to this study, the pandemic resulted in a near-total shutdown of the airline industry, which makes that time period's data very different from regular operating scenarios, which our project is targeted towards. For this reason, we reduced the dataset to only 2015-2019.

The original full weather dataset consists of weather observations, which come in a variety of report types that can be grouped into two categories. Their relevancy was determined based on research into what measurements they contain and the frequency and consistency at which they are reported. Fortunately, the relevant types (FM-15, FM-12, FM-16, and SAO) constitute 95% of all available weather observations. Types in the irrelevant category (CRN05, FM-13, SOD, SHEF, SURF, SOM, FM-14, SY-MT, SAOSP) are irregularly reported, lack the measurements of interest, and are format-incompatible with the main relevant types, so they were removed from the dataset.

A further wrangling issue arose because a percentage of rows used a backup weather station: the identifying fields for the primary station were untouched while additional identifying fields for the backup weather station were populated instead, leaving no singular column with which to determine the location (latitude, longitude, and elevation) of observations. We remedied this by (1) creating a table of unique weather stations and (2) using that table to update the primary weather station information whenever possible. A separate issue was that many weather measurement fields were represented as text, which meant the data as-is could not be used in any numeric computations; sometimes these numeric values had text qualifiers and, when missing, were represented in a variety of ways. To address these issues, measures were taken to convert such fields to numeric values and to break out their text qualifiers into separate fields as needed while standardizing the representation of missing values.

### Context (Pressure Altitude, Density Altitude and Motivation for Data Imputation & Feature Engineering)

Aircraft performance is governed by the operating conditions, which can be quantified via pressure altitude and density altitude. Pressure altitude tells pilots what altitude an aircraft's instruments show when set to the standard sea-level pressure of 29.92 inHg while ignoring temperature. Density altitude gives a fuller picture by taking into account the temperature, telling pilots the density of air. These two metrics utilize three key pieces of information: temperature, pressure, and elevation. If the air is too dense or too low, these metrics will reflect the operating conditions and flights will be delayed (or cancelled) as appropriate for safety reasons.

### Data Imputation

In the original full weather dataset, a large percentage of pressure-related fields were null — approximately 60% of `HourlySeaLevelPressure` and 40% of `HourlyStationPressure` were missing. Imputation was performed in two stages. First, in cases where one value was present but the other was missing, the present value was used to calculate the missing one, since only an elevation adjustment was needed. Second, for the remaining rows where both pressure fields were null, both fields were estimated from station pressure derived directly from elevation and temperature. For transparency, auxiliary fields were added to indicate that a value was imputed. After these procedures, pressure altitude and density altitude could finally be calculated.

A second imputation target was the wet bulb temperature field, approximately half of which was null. Missing values were imputed using the Stull formula, since wet bulb temperature is useful for downstream feature engineering.

### Feature Engineering

**Temperature features.** We engineered three temperature-based signals. Dew point spread — the difference between the dry and wet bulb temperatures — serves as a primary fog-risk indicator, since a spread of 5°F indicates that air is near saturation and likely to produce fog, which can affect visibility. Wet bulb depression (dry bulb minus wet bulb) quantifies evaporative cooling potential and icing risk. Finally, binary freezing and near-freezing flags mark conditions where precipitation is likely to freeze on contact.

**Pressure features.** Altimeter deviation from the standard 29.92 inHg captures how far current pressure deviates from normal, and a categorical pressure bucket feature encodes FAA-aligned operational thresholds (dangerously low / low / normal / high / dangerously high). Pressure tendency (rising / steady / falling) and a rapid pressure-fall flag capture approaching weather systems before they arrive.

**Visibility features.** Visibility was binned into four levels and also given a binary indicator column for low visibility.

**Wind features.** Wind direction was encoded as sine and cosine to preserve continuity across the 0°/360° boundary.

**Sky conditions and ceiling.** The elevation of the cloud ceiling, as well as whether the sky is obscured, were extracted from text codes.

**Weather type indicators.** Each weather type (Rain, Snow, Fog, Thunderstorm, Drizzle, Mist, Haze, Smoke, Hail, Ice Pellets) was broken out into its own binary indicator column.

## 2b) Flight Data: Bureau of Transportation Statistics (BTS)'s Flight Records

### Data Wrangling

The original full flights dataset spanned 2015-2021. For the same COVID-19 reasons described in the weather section, the dataset was reduced to 2015-2019. The raw data contained exact duplicate rows, which were removed via deduplication before any further processing. It also contained rows for on-time, delayed, cancelled, and diverted flights; because cancelled and diverted rows represent a different prediction problem than the one we focused on, they were dropped.

From the retained rows we built a table of unique airports using all origin and destination entries within the flight data. This airports table is leveraged alongside the table of unique weather stations to create the time-series dataset, which is done by retrieving the relevant weather data from the nearest (Haversine distance) weather station to the airport of each scheduled flight.

### Feature Engineering

**Graph features (degree-based, train-derived).** We engineered five graph features on the **training-partition flight graph** to capture network structure without leaking future information. `origin_train_departures` and `dest_train_departures` are the out-degree of each origin and the in-degree of each destination (counts of distinct departures and arrivals); high-degree nodes are the high-throughput airports that dominate the network. `origin_dest_distinct_count` records the number of **distinct destinations** served from each origin in the training partition (route-degree, or fan-out). `is_hub_origin` and `is_hub_dest` are binary flags derived from the top quantile of training-partition departure / arrival volume — a coarse, train-only centrality threshold for hub status. All five are fit on the **training partition only** and joined to validation / test as frozen lookup values, so no future-flight information leaks (Section 4).

**Temporal features.** Binary indicator flags mark weekend and late-night departures, and sine / cosine cyclical encodings preserve continuity for departure hour, day of week, month, and quarter. Departure time was also bucketed into five coarse time blocks (overnight, morning, midday, evening, night).

**Holiday / calendar features.** Binary indicator flags mark holidays and holidays whose adjacent weekend (before or after) coincides with holiday demand.

**Congestion features.** Scheduled departures were counted per origin airport in 2-hour CRS windows, producing `origin_2h_sched_departures` as a measure of how busy the airport is scheduled to be around each flight's departure. `origin_sched_count_prev2h` counts scheduled flights in the 2-hour window immediately preceding each departure, capturing the backlog the flight is departing into. `is_origin_busy_2h_before` is a binary flag derived from a training-only 90th-percentile threshold on `origin_sched_count_prev2h`.

**Target encoding features.** High-cardinality categorical fields — carrier, origin airport, destination airport, route, and departure hour — were replaced with their mean delay rate computed on the training partition only, then broadcast to validation and test with unseen groups filled by the global training mean. A 7-day rolling route delay rate was also computed using a prior-days-only lookback to capture recent route-level performance trends without same-day leakage.

**Weather features.** Each flight was joined to weather observations from its closest NOAA station falling within a 3-hour window ending 1 hour before scheduled departure — the latest point at which a prediction could realistically be made. Numeric weather fields were averaged across that window, while binary flags and categorical fields took the most recent non-null value. The five closest stations per airport were identified via a Haversine-distance cross-join, with the closest used for the primary join.

**Airport static features.** Training-partition departure counts per origin and destination airport capture long-run volume differences between airports, and hub flags mark airports whose departure counts meet or exceed the 95th percentile of the training distribution. Route diversity (`origin_dest_distinct_count`) captures how many unique destinations each origin serves, distinguishing large connecting hubs from point-to-point airports.

**Interaction features.** Several cross-feature interactions expose context-dependent delay risk. Holiday and peak flags were crossed with delay rates to capture whether busy periods are disproportionately worse at high-delay carriers or routes. Hub status was crossed with congestion counts to distinguish busy periods at major hubs from busy periods at smaller airports. Weather bins were crossed with peak-hour flags to capture whether bad weather hits harder during already-congested departure windows, and scaled weather fields were crossed with each other to expose nonlinear compound effects (for example, high wind combined with low visibility).

**Engineered index features.** A severe-weather index combines precipitation, wind, and visibility into a single composite score reflecting overall weather severity at departure time. A temperature-anomaly feature captures how far the observed temperature deviates from the historical station-month climatological norm, flagging unusually hot or cold days beyond what season alone would predict. Tail-number lag features capture whether the same aircraft's previous leg was delayed, since a late inbound aircraft directly constrains the outbound departure.

## 2c) Scaling and Standardization

Due to the nature of time-series data, feature scaling was applied carefully to avoid feature leakage. After joining the flight and weather data, six core hourly weather columns — dry bulb temperature, wind speed, relative humidity, sea-level pressure, visibility, and precipitation, plus dew point temperature — were standardized using a **rolling median and median absolute deviation (MAD)** approach. For each column, a 7-day causal lookback was computed on daily training aggregates, with the same calendar day excluded via a one-day shift to prevent any same-day leakage within the training partition. The standardization formula applied was `(value − rolling median) / rolling MAD`, producing scaled `*_scl` variants. Rather than refitting these statistics on validation or test data, the rolling median and MAD were **frozen at the last training date** and applied as fixed constants to both splits — consistent with how a production scoring pipeline would operate, where future data cannot inform the scaler.

These scaled columns serve primarily as inputs to weather interaction features rather than as direct model inputs. The raw averaged weather columns coexist alongside their scaled counterparts in the feature vector, giving the model access to both the absolute level and the standardized deviation of each weather signal.

All other feature families — target encodings, binary flags, sinusoidal time encodings, weather bins, congestion counts, airport static attributes, and graph centrality scores — were left in their natural units or encodings and not explicitly scaled. This is appropriate for tree-based models, which are scale-invariant and select splits based on rank ordering rather than absolute magnitude.

**Why StandardScaler for the deep-learning path.** MLPs converge faster and more stably on zero-mean / unit-variance inputs because gradient updates land on a similar scale across features and weight initialization assumes that distribution. We therefore use **StandardScaler** for the MLP / sklearn track. The tree-based path (GBT) reads the **assembler-only** raw inputs because tree splits are scale-invariant and benefit from the original distribution shape; **LR / RF** use the StandardScaler manifest by convention so coefficients and feature importances live on a common scale.

**Temporal split details.** The 60 / 20 / 20 train / validation / test split is computed on **distinct calendar dates** (`_sort_date`) rather than on row counts, so each split spans a contiguous date range and the validation / test horizons sit strictly **after** every training row in real time.

#### Feature Correlations

A correlation heatmap of the top 20 features (ranked by absolute **Spearman rank correlation** with our target, `DEP_DEL15`) reveals that target-encoded delay rates (`hour_delay_rate`, `carrier_delay_rate`, `route_delay_rate`) and temporal features (`dep_hour`, evening/morning peak flags) exhibit the strongest individual associations with the target. Several engineered interaction terms also appear among the top 20, confirming that the cross-features carry predictive signal beyond their constituent parts.

*A note on method.* `DEP_DEL15` is a **binary** outcome, so a **Pearson** correlation — which assumes linear relationships between continuous variables — is not the right association measure. We use **Spearman rank correlation**, which (a) handles binary variables cleanly through tied ranks, (b) does not require a linearity assumption between the feature and the target, and (c) keeps a single, business-readable color scale across the entire matrix.

Feature Correlations (Spearman)

#### Inbound aircraft cascade (`tail_prev_delay`)

A delayed inbound leg directly squeezes the next departure on the same airframe — turnaround time gets compressed and any gate / crew slack disappears. We capture this with `tail_prev_delay`, a per-aircraft lag of `DEP_DEL15`: for every flight, look at the previous leg flown by the **same tail number** (sorted by date and scheduled departure) and emit `1` if that prior leg was delayed ≥ 15 minutes at departure, else `0` (the first leg of a tail in the data is filled as `0`). The feature is **history about the previous segment**, not a forward-looking signal, so it stays leakage-clean (Section 4).

On the **training split** (15.1M flights with `tail_prev_delay = 0`, 3.6M with `tail_prev_delay = 1`), the base-rate departure-delay frequency rises sharply when the prior leg was already delayed:

| `tail_prev_delay` | Delay rate (`DEP_DEL15 = 1`) | Flights |
|:-----------------:|:----------------------------:|--------:|
| 0 (prior leg on time / no prior leg in data) | **17.4%** | 15,139,881 |
| 1 (prior leg delayed ≥ 15 min)               | **28.0%** | 3,559,118 |

That is a **+10.6 percentage-point** absolute lift (a **+60.6 %** relative increase) on a single binary indicator — the kind of high-information, low-leakage signal that materially improves every downstream model that consumes it (Section 6c).

## 3a) Motivation

The client specifically requested a neural network implementation to complement the classical ML models (LR, RF, GBT). Multilayer perceptrons can capture nonlinear feature interactions through learned hidden representations — potentially discovering complex patterns in the 99-dimensional feature space that linear models miss and that tree ensembles approximate only through axis-aligned splits.

We pursued two complementary implementations: **PySpark's `MultilayerPerceptronClassifier`** for full-scale distributed training on the complete dataset, and **scikit-learn's `MLPClassifier`** for a rigorous ablation study on a stratified subsample. This dual approach was necessary because PySpark's MLP API is limited: it supports only the L-BFGS optimizer (no Adam), exposes no per-epoch training loss, offers no regularization (L1/L2) or dropout, and provides no early stopping mechanism. The scikit-learn complement fills each of these gaps.

## 3b) PySpark `MultilayerPerceptronClassifier`

**Preprocessing pipeline**: Tabular **LR** and **RF** consume Phase 2.4 **Notebook A** ML parquets (`phase2_df_undersample_*_ml`) with a train-fitted **StandardScaler** on engineered numerics (with `*_scl` pass-through where defined). **GBT** uses Phase 2.4 **Notebook B** (assembler-only). The **Spark MLP** stack is a **separate pipeline**: it loads Phase 3.3 **feature-engineered** parquets (`df_train_phase3.parquet` / val / test), **fits its own** `VectorAssembler` + `StandardScaler` (+ optional PCA), then applies **2:1** majority undersampling for training — so MLP preprocessing is not the same serialized pipeline as Phase 2.4 Notebook A, even though both use StandardScaler in spirit.

**Architecture search**: We experimented with three architectures, each using sigmoid hidden activations and softmax output (the only options available in PySpark's MLP). The fitted vector dimension after scaling was **104** (manifest / parquet alignment in the MLP notebook).

| ID | Layer Configuration | Hidden Units | Preprocessing | Rationale |
|----|---------------------|-------------|---------------|-----------|
| MLP-A | [104, 64, 2] | 64 | StandardScaler (in-notebook) | Shallow baseline — single hidden layer, tests whether a simple nonlinear mapping suffices |
| MLP-B | [104, 128, 64, 2] | 128 + 64 | StandardScaler (in-notebook) | Two hidden layers — deeper representation learning with progressive dimensionality reduction |
| MLP-C | [104, 256, 128, 2] | 256 + 128 | StandardScaler (in-notebook) | Wider first layer — maximum capacity to capture complex interactions |

**Optimizer**: PySpark's MLP supports only `l-bfgs` (a quasi-Newton method) and vanilla gradient descent. We used L-BFGS, which computes an approximation to the inverse Hessian and typically converges faster than first-order methods for moderately-sized networks. However, L-BFGS requires storing the full gradient in memory and does not support minibatch training, making it less scalable than Adam for very large networks.

**Class imbalance handling**: Phase 3.3 engineered data / undersampled train, imbalance handled by train undersampling in FE. The MLP notebook additionally applies **2:1** undersampling on the scaled training frame (retaining all delayed flights and downsampling on-time flights) before fitting.

**PCA experiment**: We also tested dimensionality reduction via PCA (retaining 95% of variance), reducing the input to **55** principal components in our run before the final hidden layers. This tests whether the feature space contains redundant dimensions whose removal improves generalization or training speed.

#### Optimizer Comparison

We compared three optimizers on identical data and architecture. **Adam** (adaptive moment estimation) maintains per-parameter learning rates using first and second moment estimates of the gradient, and generally converges faster and more reliably for neural networks. **L-BFGS** (limited-memory BFGS) is a quasi-Newton method that approximates the inverse Hessian and is PySpark's only option for MLP; it is effective for smaller networks but does not support minibatch training. **SGD** (stochastic gradient descent) is a first-order method with a fixed learning rate — the simplest of the three but also the slowest to converge without careful tuning.

## 3d) Limitations and Future Directions

The primary limitation of both our PySpark and scikit-learn MLP implementations is the absence of **dropout regularization**. Dropout randomly deactivates neurons during training, forcing the network to learn redundant representations that generalize better. Neither PySpark's `MultilayerPerceptronClassifier` nor scikit-learn's `MLPClassifier` supports dropout; implementing it would require a **PyTorch** or **TensorFlow** backend with distributed training via Horovod or PyTorch's `DistributedDataParallel`.

Several related capabilities are also unavailable in the current stack. **Batch normalization**, which stabilizes training and allows higher learning rates by normalizing activations between layers, is not available in either PySpark or sklearn MLP. **Custom activation functions** are similarly constrained: PySpark MLP is locked to sigmoid hidden activations and softmax output, so ReLU — which avoids the vanishing gradient problem — is not supported. **Learning rate scheduling** is partially mitigated by Adam's adaptive rates, but explicit schedulers such as cosine annealing and warm restarts are unavailable. Finally, PySpark's **fixed solver** leaves only L-BFGS as a practical choice at scale, so Adam (generally superior for deep networks) is not available on that track.

**Future work:** A PyTorch implementation running on GPU-enabled clusters would unlock deeper architectures (3+ hidden layers), ReLU activations, dropout, batch normalization, and cosine-annealed learning rate scheduling — each of which could meaningfully improve F2 beyond what the current constrained MLP achieves.

# 4) Data Leakage Discussion

### 4a. Definition

Data leakage occurs when information that would not be available at prediction time — typically future outcomes, post-event measurements, or test-set statistics — is inadvertently incorporated into the model's training data. Leakage produces artificially inflated performance metrics that do not generalize to real-world deployment, because the model has effectively been given access to the answer it is trying to predict.

### 4b. Hypothetical Example

Consider including `DEP_DELAY` (actual departure delay in minutes) as a feature to predict `DEP_DEL15` (binary: delayed ≥ 15 minutes). Since `DEP_DEL15` is derived directly from `DEP_DELAY` (it equals 1 if and only if `DEP_DELAY` ≥ 15), any model would achieve near-perfect accuracy — not because it learned to predict delays, but because the answer was embedded in the input. This is the most extreme form of leakage: using the outcome to predict the outcome.

More subtle forms include using `WHEELS_OFF` time (which is observed after departure), actual taxi-out duration, or any measurement that is recorded *after* the scheduled departure time.

### 4c. Pipeline Leakage Audit

We systematically audited every stage of our pipeline for potential leakage vectors:

| Pipeline Stage | Leakage Risk | Mitigation |
|---------------|-------------|------------|
| Feature selection | Post-departure fields (`DEP_TIME`, `WHEELS_OFF`, `TAXI_OUT`, `DEP_DELAY`, actual arrival data) encode outcome information | Dropped 25+ post-departure columns before any modeling (see `LEAKY` list in Phase 3.3 feature engineering notebook) |
| Target encoding | `carrier_delay_rate`, `route_delay_rate`, etc. are computed from labeled data and could leak if computed on the full dataset | Computed on **training split only**; values broadcast to validation/test via lookup join |
| Congestion threshold | `is_origin_busy_2h_before` uses a threshold that could reflect val/test data if computed globally | Threshold is the **train-only 90th percentile** of `origin_sched_count_prev2h` |
| Temporal split | Random shuffling would mix future flights into training, violating temporal ordering | Deterministic 60/20/20 split by `(MONTH, DAY_OF_MONTH)` — no future data contaminates training |
| Class weights | Weight computation from label distribution could reflect val/test if computed globally | Weights computed on **training split only** |
| StandardScaler (tabular + MLP) | Fit statistics (mean, std) must not include validation/test observations | Fit on **training data only**; transform applied to val/test with frozen statistics |
| PCA | Principal components must not encode val/test variance structure | Fit on **training data only** |
| Graph degree / hub flags | Counts and hub thresholds computed from the full dataset would leak val/test traffic into training inputs | `origin_train_departures`, `dest_train_departures`, `origin_dest_distinct_count`, and the `is_hub_origin` / `is_hub_dest` thresholds are all fit on the **training partition only** and broadcast to val/test as frozen lookup values |
| Threshold tuning | Tuning the decision threshold on test data would invalidate blind evaluation | Tuned on **validation set only**; test threshold is frozen |
| Blind test evaluation | Any use of test data for decisions would contaminate evaluation | 2019 test data evaluated **once** with frozen hyperparameters and threshold |


### 4d. Cardinal Sins of Machine Learning

We verify that our pipeline does not violate any of the cardinal sins of machine learning. First, there is **no test data in training or hyperparameter selection** ✓ — the 2019 blind test set was never used during model training, threshold tuning, or hyperparameter search, and all tuning decisions were made on the validation set. Second, **no future data leaks into past predictions** ✓ — our temporal split ensures that training data (2015–2018, selected calendar days) never includes flights from the validation or test partitions, and features are computed only from information available before each flight's scheduled departure. Third, **no outcome variables are used as features** ✓ — all post-departure columns (actual departure time, taxi-out, wheels-off, actual delays, arrival data) were explicitly dropped before feature engineering. Finally, **cross-validation respects temporal structure** ✓ — grid-search cross-validation uses the built-in PySpark `CrossValidator` with `(MONTH, DAY_OF_MONTH)`-based fold assignments rather than random shuffling, preserving temporal ordering within folds.

### 5c. Loss Functions

**Logistic Regression** — Binary cross-entropy with elastic net regularization:

> L_LR = −Σᵢ [ yᵢ log(p̂ᵢ) + (1 − yᵢ) log(1 − p̂ᵢ) ] + λ [ α ‖w‖₁ + ((1−α)/2) ‖w‖₂² ]

where λ is `regParam` and α is `elasticNetParam` (0 = pure L2/Ridge, 1 = pure L1/Lasso).

**Random Forest** — Gini impurity for node splits:

> G = 1 − Σₖ₌₁..ₖ pₖ²

where pₖ is the proportion of class k in the node. The forest aggregates predictions across 100 independent trees trained on bootstrap samples with feature subsampling.

**Gradient-Boosted Trees** — Log loss (binary cross-entropy) with sequential additive boosting:

> L_GBT = Σᵢ ℓ( yᵢ, Fₘ₋₁(xᵢ) + η · fₘ(xᵢ) )

where Fₘ₋₁ is the ensemble after m−1 rounds, fₘ is the new tree fit to the negative gradient (pseudo-residuals), and η is the step size (learning rate).

**Multilayer Perceptron** — Binary cross-entropy (no explicit regularization in PySpark; L2 via `alpha` in sklearn):

> L_MLP = −Σᵢ [ yᵢ log(p̂ᵢ) + (1 − yᵢ) log(1 − p̂ᵢ) ]

### 6b. Cross-Phase Comparison: Did Feature Engineering Help?

Comparing **Phase 2** (minimal stack, **13** numeric inputs) to **Phase 3** (engineered pipeline, **99** feature *families* and **~103–104** columns in the Phase 2.4 ML manifests) shows **clear gains on the five-year splits**, with the caveat that Phase 2 and Phase 3 are not identical experimental clones: Phase 3 uses the Phase 3.3 FE parquets, Phase 2.4 export paths (StandardScaler vs assembler-only), and updated imbalance handling (e.g., undersampled training for RF/GBT flows; LR with class weights where configured).

On **AUROC (validation)**, the tabular models move as expected if the new features carry signal: **GBT** rises from the Phase 2 default-style baseline (**0.6726** in the Phase 2 report) to **0.6954** under Phase 3 grid search with a tuned decision threshold (**0.20**). **LR** moves from the Phase 2 grid-search reference (**0.6551**) to **0.6624** in Phase 3, a modest lift consistent with a still-mostly-linear boundary that only partially exploits interaction-rich columns. **RF** is not apples-to-apples with a single Phase 2 row: Phase 2 included a **class-weighted** configuration (**0.6681** AUROC in the Phase 2 table), whereas Phase 3 emphasizes **undersampled training** with **threshold = 0.4** (**0.6741** AUROC). The point is not “RF went down,” but that **the training recipe changed**; under Phase 3’s recipe, RF **AUROC is competitive** and sits between LR and GBT as expected.

**F_2** (beta = 2) is only fully informative once you compare within Phase 3, where it is reported consistently. On **validation**, LR / RF / GBT land in a tight band (**0.5077–0.5226**), with **recall on the order of 0.81–0.86** at the tuned thresholds—i.e., the models **catch most delayed flights**, trading off **precision in roughly the 0.20** range (many false alarms). On the **held-out test** split, the same three models’ blind **F₂** values cluster near **0.52–0.527**, which is the more honest “did the pipeline generalize?” check than validation alone.


### 6c. Model Comparison

On the **five-year** Phase 3 splits (Phase 2.4 ML exports; validation metrics on the engineered splits), **GBT** achieves the strongest **validation** performance among the classical tabular models: **F₂ = 0.5226** and **AUROC = 0.6954** at tuned threshold **0.20**. **RF** is next (**F₂ = 0.5147**, **AUROC = 0.6741**, threshold **0.4**), with the **highest validation recall** (**0.8544**). **LR** is slightly lower on **F₂** (**0.5077**) and **AUROC** (**0.6624**), with **recall 0.8098**—still delay-heavy under F₂-oriented tuning, but less aggressive than the tree models.

**Generalization (blind test).** Test **F₂** values are **tight rather than identical**: **LR 0.5219**, **RF 0.5228**, **GBT 0.5267** (same validation-tuned thresholds). This supports the interpretation that, once thresholds are chosen to optimize delayed-flight capture, **the tabular models land in a similar operating regime on held-out data**, with remaining spread driven by model form and calibration rather than a large validation–test collapse.

**Tradeoffs (tabular).** **LR** is typically the **most interpretable** linear baseline and the natural reference for coefficient-style explanations; it **underperforms** tree models on **AUROC** and **F₂** here, consistent with limited exploitation of engineered nonlinear interactions. **RF** benefits from **independent trees** that make training more parallel-friendly than boosting in principle; in this project RF is trained on **Phase 3.3 undersampled** rows **without** `weightCol` (to avoid **double** imbalance correction versus undersampling), and although **GBT** remains the strongest discriminator, RF is typically cheaper per comparable quality tier because boosting is inherently sequential. **GBT** delivers the best **AUROC** and the best validation **F₂** of the three LR/RF/GBT runs; it uses the **assembler-only** 104-d export path and **does not** use `weightCol` in Spark — imbalance is handled via **training design + thresholding** rather than per-row weights.

**MLP (PySpark).** The Spark MLP stack fits an in-notebook **StandardScaler**, applies a **2:1** majority undersample for training, and evaluates on **full** validation/test. Among A/B/C at the default **0.50** threshold, **MLP-A** **[104, 64, 2]** is strongest on validation **F₂** (**0.4554**; **AUROC 0.6553**). **Threshold tuning** on validation raises **MLP-A** to **F₂ = 0.5128** at **0.30**, with blind test **F₂ = 0.5212** at that frozen threshold. **PCA (55 components)** reduces validation **F₂** to **0.3960** in our run, indicating the **104-dimensional** scaled inputs were preferable for this dataset and model class.

### 6d. Threshold Analysis

Binary classifiers emit a **delayed** probability; the **decision threshold** maps that score into a hard prediction. The default **0.5** cutoff is a poor default under **severe imbalance** when the business loss is asymmetric: missing a delay is typically far more costly than a false alarm. Because our primary score is **F₂** with **β = 2** (recall weighted twice relative to precision in the standard F_β definition, matching the printed metric headers in our modeling notebooks), we **sweep thresholds on the validation split**, pick the cutoff that **maximizes validation F₂**, and only then apply that frozen threshold to **blind test** reporting.

The table below shows the **GBT validation threshold sweep** (grid-selected GBT, Phase 3.3 undersampled training policy, sweep grid `THRESHOLDS = [0.10, …, 0.50]`). It illustrates the expected precision–recall trade curve as the threshold moves from aggressive delay-catching (low threshold) toward conservative 0.5 behavior.

| 0.15 | 0.5035 | 0.1721 | 0.9712 |
| 0.20 | 0.5147 | 0.1867 | 0.9176 |
| **0.25** | **0.5182** | **0.2005** | **0.8582** |

**Selection (sweep snapshot).** Under the grid-search GBT shown above, the sweep maximum sits at **threshold = 0.25** (**F₂ = 0.5182**, **recall 0.8582**, **precision 0.2005**). After the post-presentation re-run, a stronger grid-search GBT lands at **threshold = 0.20** with **validation F₂ = 0.5226**, **AUROC = 0.6954**, **precision 0.2130**, **recall 0.8209** (Section 6a) — a small upward shift in optimal cutoff that follows from the higher AUROC. The qualitative shape of the curve — F₂ peaking well below 0.50 because aggressive recall is rewarded under β = 2 — holds in both runs. At **0.50**, F₂ collapses to **0.3677** with **recall 0.3907**, i.e., the model would miss most delays — unacceptable for a delay-alerting use case.

**Related thresholds (same methodology).** **RF** is reported at threshold **0.4** and **Spark MLP-A** threshold tuning selects **0.30** (validation F₂ = 0.5128 for that MLP run). These values are likewise **validation-tuned** and should be interpreted as **operating points**, not intrinsic model quality.

**Operational interpretation.** At the latest GBT operating point (**threshold = 0.20**), roughly **one in five** flagged flights is a true delay (**precision ≈ 0.21**), while **recall remains high** (**≈ 0.82**). Whether that false-positive burden is acceptable is a **business-cost** question; the modeling contribution is to make the trade explicit and to choose the cutoff that aligns with the team’s **F₂** objective rather than default **0.5** accuracy thinking.

### 6e. Gap Analysis

Even after scaling to the **five-year** engineered splits (**~14.5M** train rows in the Phase 2.4 export policy) and tuning decision thresholds for **F₂** (β = 2), the operating point we report for **GBT** (**threshold = 0.20**) still implies a **precision–recall trade**: validation **F₂ ≈ 0.523**, **recall ≈ 0.82**, and **precision ≈ 0.21** (Section **6d**). On the **blind 2019** split, **LR / RF / GBT** remain in a tight **F₂** band near **0.52–0.527** (Section **6a**), so the gap is less “validation vs. test collapse” and more “what systematic errors remain at this operating point?”

**Where the model tends to fail.** First, **operational delays without strong tabular cues** — when a delay is driven by **mechanicals, crew, gate turnaround, or airline-specific ops** that are not encoded in BTS + LCD-style weather and schedule features — can still be predicted as on-time because the strongest learned signals are **temporal delay rates**, **target-encoded route/carrier/hour behavior**, and **weather severity bins** (Sections **2d**, **6c**, **6e**). Second, **sparse or unstable target encodings** such as `carrier_delay_rate`, `route_delay_rate`, and related training-only means can be **noisy** for **low-volume** carriers and routes and can behave poorly for **new** market pairs that were rare in **2015–2018** training. Third, the graph structure is **limited to degree-only features**: we capture **local** congestion (`origin_sched_count_prev2h`, etc.) and **degree-based** structural importance (`origin_train_departures`, `dest_train_departures`, `origin_dest_distinct_count`, plus the thresholded `is_hub_origin` / `is_hub_dest` flags), all fit on the training partition only, while spectral measures we did **not** compute include **PageRank** (importance weighted by neighbors' importance), **betweenness centrality** (criticality on shortest paths), and **dynamic / live** downstream-hub state; adding those — particularly a leakage-respecting monthly PageRank or a real-time inbound-aircraft / crew constraint signal — is the most direct next move for the **cascade** scenarios where the origin looks fine but the system is already stressed elsewhere. Fourth, there is a **precision ceiling under imbalance + recall-first tuning**: with **~20%** positive prevalence (Phase **2** references) and thresholds chosen to **catch delays**, **precision near ~0.20** is an expected companion to **high recall** unless new features materially separate true positives from false positives.

**What would most likely move the needle next.** The most promising next step is **inbound / tail-circuit realism (`tail_prev_delay`)**. We engineer `tail_prev_delay` by sorting all rows so that the previous leg is the earlier real-time event for each aircraft (`Window.partitionBy('TAIL_NUM').orderBy(_sort_date, CRS_DEP_TIME)`) and taking `lag(DEP_DEL15, 1)` over that window, cast to `double` and `fillna(0.0)` for the first leg of each tail. The flag is **history about the previous segment** (`1` = the same aircraft's prior leg was delayed ≥ 15 min at departure; `0` = either the prior leg was on time or no prior leg exists in the window), so it adds turnaround-cascade information **without** violating the leakage discipline of Section 4. The remaining lever is **richer same-day tail-circuit telemetry** — multiple-leg lookback, scheduled-vs-actual turnaround minutes, or live inbound ETA — which is the most direct next move for turn-constrained flights. A complementary extension is **forward-looking weather**: **TAF** (and similar products) would complement **historical LCD** windows that end **1 hour before** scheduled departure. Finally, a **stronger sequence / DL stack** is worth pursuing: the **Transformer** row in the consolidated experiment table reaches a competitive **blind** **F_β** relative to GBT, and a production path likely needs **GPU training**, careful **tokenization**, and the same **leakage discipline** as tabular models (Section **4**).

**A note on train-set metrics.** The five-year LR / RF / GBT runs were trained on **~14.5 M** rows; re-evaluating those models on the full training split would have added tens of thousands of seconds of cluster time on `Team_1_2` (non-GPU) without changing the operating-point story. We rely on three convergent overfit checks instead. First, the MLP **training-vs-validation curves** (Section 5a / 6c) show the val-loss tracking the train-loss within a small constant gap and an early-stop trigger before divergence. Second, the **tight validation-vs-blind-test gap** for tabular models (~0.518 → ~0.524 on F₂ for GBT after the post-presentation re-run) is itself evidence against severe overfit, since the blind 2019 split was never consulted during training. Third, the **feature-family ablation** (Section 6c) is consistent with the gain coming from real signal, not memorization, since dropping any one family degrades F₂ on the held-out split.

# 7) Conclusion

We framed **May(De)lay** as **binary classification** for **15+ minute departure delay** (`DEP_DEL15`) using information aligned to a **decision time about one hour before** scheduled departure: **BTS** operational records joined to **NOAA LCD** weather, engineered into **eight feature families** totaling **104** numeric inputs on the Phase **3** manifest (`phase3_final_num_cols.json`), evaluated on **2015–2019** data with a **temporal split** and an audited **leakage** posture (Sections **2**, **4**).

Across phases, the story is **iterative improvement through data scale + feature engineering + thresholding**: Phase **2** established strong baselines on a smaller engineered footprint; Phase **3** expanded interactions, congestion, and weather structure, and moved the main classical comparisons to the **five-year** Phase **2.4** ML exports (StandardScaler path for **LR/RF**, assembler-only path for **GBT**; Section **5**).

**What worked best in our reported tables (tabular + Spark MLP merge in Section 6a).** Among **LR / RF / GBT**, **GBT** achieved the strongest **validation** discrimination and **F₂** in the merged Phase **3** comparison (**F₂ = 0.5226**, **AUROC = 0.6954** at **threshold = 0.20**). On the **blind** test split, **LR**, **RF**, and **GBT** cluster tightly in **F₂** (**~0.5219–0.5267**), indicating **similar operating regimes** once thresholds are tuned for recall-heavy **F₂**—not a single model “dominating” out-of-sample by a large margin.

**Deep learning (client requirement).** **PySpark MLP** favored the **shallower** **MLP-A** configuration at the default **0.50** threshold and benefited substantially from **validation threshold tuning** (Section **3b** / **6a**). **sklearn MLP** on a stratified subsample supported rigorous **optimizer / L2 / feature-family** analysis (Section **3c**), while noting API limits (Section **3d**). Separately, the consolidated experiment table includes **PyTorch MLP** sweeps and a **feature-token Transformer** run (Section **5a**) that reaches a competitive **blind** **F_β** relative to the best tabular models — confirming that, at the data scale available on a non-GPU cluster, simpler architectures sit on the same Pareto frontier as deeper ones once thresholds are tuned for F₂.

**Practical interpretation.** At the **GBT** tuned point discussed in Section **6d**, the system is tuned to **miss fewer delays** at the cost of **many alerts** (**precision ~0.21**, **recall ~0.82** on validation). Whether that trade is acceptable is a **business** question; the modeling contribution is to **make the trade explicit** and select thresholds against **F₂** rather than default **0.5** thinking.

**Future work** matches the largest remaining gaps: **richer operational telemetry** (inbound delay/crew/maintenance where available), **forecast weather** (**TAF**) beyond retrospective LCD windows, and **modern DL training** (dropout / batch norm / attention) with **distributed** tooling—while preserving the same **temporal + leakage** constraints that make the current numbers credible.

# 8) Data lineage, feature catalog, and EDA map

**May(De)lay** predicts **binary departure delay** (`DEP_DEL15`: **≥ 15 minutes late**) using information aligned to about **one hour before** scheduled departure. The project combines **BTS-style flight records** with **NOAA LCD-style weather**, joined at flight grain, evaluated on **2015–2019** domestic U.S. operations (with pandemic-era years excluded per team policy), using a **temporal split** and a documented **leakage audit** in the final report.

This endcap summarizes how data move from raw sources to model-ready tables, how the final feature vector is organized, and where exploratory and associational evidence appears across **Phase 1–3** deliverables.

---

## Data lineage (project arc)

### Phase 1 — ingestion, cleaning, and foundations

For **flights**, Phase 1 focused on cleaning (for example, deduplication and scoping away cancelled or diverted rows where applicable) alongside airport and schedule context. The **weather** pipeline applied LCD-style filtering to operationally relevant report types, performed numeric parsing, repaired station identity, and imputed the pressure and wet-bulb fields that downstream aviation-style features depend on. A **bridge** step built the airport ↔ nearest-station mapping (Haversine) so each flight can attach a **causal weather window** before the decision time. The phase **deliverables** were analysis-ready tables, missingness and distribution EDA, and a reproducible path to Phase 2 scale-up.

### Phase 2.2 — flight–weather join (one row per flight)

Phase 2.2 implements the flight–weather join as a one-row-per-flight table. The **map** step links each origin to a **primary nearest weather station** (e.g. `closest_station_1` / `wx_station`). The **window** step filters weather rows by date around the flight and by timestamps in **`[weather_window_start, weather_window_end]`** relative to scheduled departure. The **aggregate** step then runs `groupBy` on the flight keys: **`avg_*`** for numeric weather (and numeric engineered weather carried forward), and **`last_*`** for binary and categorical engineered fields. The resulting **artifact** is `kenzie_full_df_joined.parquet` (paths under team `BASE_GROUP`), produced by the **source notebook** `Phase2.2_Flight_Weather_Join_Pipeline.ipynb`.

### Phase 2 — baselines

Phase 2 worked with a smaller engineered footprint under a temporal validation design, ran imbalance experiments, trained **LR / RF / GBT** baselines, and produced the midpoint-report EDA (delay-by-hour, prevalence, and early correlations).

### Phase 3 — engineered scale-up and numeric contract

In **Phase 3.3 — feature engineering**, we applied leakage-aware drops; used a **60/20/20** split on **`(MONTH, DAY_OF_MONTH)`**; built CRS-only **holiday** and **congestion** features; added **target encodings** and **7-day route context** (below); introduced **weather bins** and **scaled** hourly interactions; computed **airport statics**; layered extended **A–E** interactions; applied an optional **train-only majority undersample**; and emitted **`phase3_final_num_cols.json`** alongside `df_{train,val,test}_phase3.parquet`.

## Feature families (taxonomy) — final engineered stack

| Family | Intent | Train-only / leakage discipline | Depends on |
|--------|--------|----------------------------------|------------|

## Encoding and train-derived rate features

This project implements **more than a single formula**: it combines **mean target encoding** with a **global-mean back-off**, a **causal rolling route statistic**, and (later) **algebraic combinations** of encoded columns.

| Mechanism | What the notebook does | Why |
|-----------|-------------------------|-----------------------------------|
| **Global mean** | `global_mean = mean(DEP_DEL15)` on **`df_train_clean`** | Single **prior delay rate** for **missing joins** after `left` joins (rare carriers, new routes in val/test, sparse daily history). Stabilizes high-cardinality groups with few samples—same role as a simple **Bayesian back-off** without adding a separate smoothing hyperparameter. |
| **Mean target encoding (per key)** | `make_target_enc`: `groupBy(group_col).agg(mean(TARGET))` on **`df_train_clean`** for carrier, origin, dest, route, hour | Notebook motivates this as **one numeric signal per high-cardinality categorical** instead of exploding **airport / route** cardinality into one-hot (see EDA: *“Large airport cardinality motivates target encoding later instead of one-hot.”*). |
| **Apply + fill** | `apply_enc`: `left` join encoding table, then **`fillna(global_mean)`** on the new column | Ensures **val/test never refit** encodings; unseen groups inherit the **train-wide** delay rate rather than nulls or arbitrary zeros. |
| **7-day rolling route rate** | `route_delay_rate_7d`: daily `mean(DEP_DEL15)` by **`ROUTE` × calendar day** on train, then **average of prior days’** `_daily_rate` with **`b.fl_date_int < a.fl_date_int`** and **`b.fl_date_int >= a.fl_date_int - 7`**; join to all splits; **`fillna(global_mean)`** | Captures **recent route performance** without using **same-day** outcome in the average (strict inequality on past dates). In the notebook’s **0.5% train Pearson** ranking, **`route_delay_rate_7d`** sits next to **`route_delay_rate`**, supporting incremental signal beyond the **lifetime** route mean. |
| **Products of encodings** | e.g. `evening_peak_x_route_delay`, `carrier_x_origin_delay`, … | Notebook **Takeaways** and full-candidate correlation show **peak × rate** and **carrier × origin** terms rank highly—**multiplicative** structure catches **context-dependent** delay risk that additive rates miss. |
| **Residual / contrast** | `route_carrier_residual = route_delay_rate - carrier_delay_rate` | Highlights **route-specific** hardness **relative to** the carrier’s network-wide baseline (Block D residual contrast). |
| **Train-only median encoding** | `temp_anomaly_station_month`: **`percentile_approx(avg_HourlyDryBulbTemperature, 0.5)`** by **`ORIGIN`×`MONTH`** on **train**, then deviation on val/test | Notebook explicit **leakage note:** same **train-only spirit** as target encodings—flags **unusual temperature** for that station-month vs its own climatology. |

## Appendix — `phase3_final_num_cols.json` build order

The numeric manifest is assembled in a fixed order. First, `_flight_info` contributes `QUARTER`, `MONTH`, `DAY_OF_MONTH`, `YEAR`, `DISTANCE`, `CRS_DEP_TIME`, and `dep_hour`. Next, `_time_features` contributes peaks, weekend and late-night flags, and cyclical encodings. Then `PHASE3_EXTRA_NUM_COLS` adds holiday, congestion, the five `wx_*_scl` interactions, `wx_dewpoint_daily_anom`, airport statics, and `tail_prev_delay`. After that, `_target_enc` contributes **`carrier_delay_rate`, `origin_delay_rate`, `dest_delay_rate`, `route_delay_rate`, and `hour_delay_rate`** (five names per the notebook list). `_weather_bins` contributes `precip_bin`, `wind_bin`, and `vis_bin`. `_wx_feat` contributes the six core `avg_Hourly*` columns plus the other non-`_scl` `avg_Hourly*` columns. Finally, `EXTENDED_INTERACTION_COLS` contributes whichever of the A–E names are present on the frame.

# 9) Future work

Our first near-term priority is **tail-centric propagation**: extending the same **chronological tail window** idea as today’s prior-leg delay flag into **multi-hop or rolling summaries** of past delay on the **same tail**, with explicit rules for **first leg / missing tail** and for **what is known at scoring time**. A stretch-goal extension is **tail graph / PageRank-style ideas**: a **train-only** graph fit (tail or leg graph, edge weights from schedule or delay), **centrality scores**, and then joining to val/test without refitting on their labels — with documented **observability** relative to the prediction timestamp. Finally, a **feature contract** discipline must hold regardless of which extension lands first: any new column must be added to the **Phase 3 numeric manifest** and kept consistent with **Spark vs deep-learning** consumers so train/serve columns stay aligned.
